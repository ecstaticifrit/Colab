MNIST

```python
data = keras.datasets.mnist
data
```

```python
df = data.load_data()
df
```

```python
(X_train, y_train),(X_test, y_test) = df
```

```python
X_train_flat = X_train.reshape(-1, 28*28)
X_test_flat = X_test.reshape(-1, 28*28)
```

```python
names = ['sigmoid','softmax', 'tanh', 'PReLU', 'ELU', 'selu']
```

```python
for name in names:
  model = Sequential([
                    
                    Dense(units=10,
                          input_shape=(784,),
                          activation=name),
                    
                    #Dense(units=100,
                     #     activation='sigmoid')
  ])

  model.compile(
      optimizer='adam',
      loss='sparse_categorical_crossentropy',
      metrics=['accuracy']
  )

  A = model.fit(X_train_flat, y_train, epochs=10)
  y_pred = model.predict(X_test_flat)
  print("accuracy on using ", name)
  model.evaluate(X_test_flat, y_test)
  print(' ')
```

Wisconsin Breast Cancer

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/wisconsin breast cancer.csv")
```

```python
y = df["diagnosis"].values

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(y)

X = df.drop(labels = ["diagnosis", "id"], axis=1)
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
print("Shape of training data is: ", X_train.shape)
print("Shape of testing data is: ", X_test.shape)
```

```python
names = ['sigmoid','softmax', 'tanh', 'PReLU', 'ELU', 'selu']
for name in names:
  model = Sequential([
                    
                    Dense(units=10,
                          input_shape=(31,),
                          activation=name),
                    
                    #Dense(units=100,
                     #     activation='sigmoid')
  ])

  model.compile(
      optimizer='adam',
      loss='sparse_categorical_crossentropy',
      metrics=['accuracy']
  )

  A = model.fit(X_train, y_train, verbose=1, epochs=20, batch_size=64, validation_data=(X_test, y_test))
  print("accuracy on using ", name)
  y_pred = model.predict(X_test)
  y_pred = (y_pred > 0.5)
  model.evaluate(X_test, y_test)
  print(' ')
```

Iris

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/Iris.csv")
```

```python
X=df.iloc[:,1:5].values
y=df.iloc[:,5].values
```

```python
df.loc[df["Species"]=="Iris-setosa","Species"]=0
df.loc[df["Species"]=="Iris-versicolor","Species"]=1
df.loc[df["Species"]=="Iris-virginica","Species"]=2
```

```python
X_normalized=normalize(X,axis=0)
```

```python
total_length=len(df)
train_length=int(0.8*total_length)
test_length=int(0.2*total_length)

X_train=X_normalized[:train_length]
X_test=X_normalized[test_length:]
y_train=y[:train_length]
y_test=y[test_length:]

print("Length of train set x:",X_train.shape[0],"y:",y_train.shape[0])
print("Length of test set x:",X_test.shape[0],"y:",y_test.shape[0])
```

```python
y_train=np_utils.to_categorical(y_train,num_classes=3)
y_test=np_utils.to_categorical(y_test,num_classes=3)
print("Shape of y_train",y_train.shape)
print("Shape of y_test",y_test.shape)
```

```python
names = ['sigmoid','softmax', 'tanh', 'PReLU', 'ELU', 'selu']
for name in names:
  model = Sequential([
                    
                    Dense(units=3,
                          input_dim=4,
                          activation=name)
                    
                    #Dense(units=100,
                     #     activation='sigmoid')
  ])

  model.compile(
      optimizer='adam',
      loss='categorical_crossentropy',
      metrics=['accuracy']
  )

  model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=20,epochs=10)
  print("accuracy on using ", name)
  prediction=model.predict(X_train)
  length=len(prediction)

  y_label_1=np.argmax(y_train,axis=1)
  predict_label_1=np.argmax(prediction,axis=1)

  accuracy_train=np.sum(y_label_1==predict_label_1)/length * 100 
  print("Train data Accuracy = ",accuracy_train )
  print(' ')
```
